```python
data2 = pd.read_csv("/content/1. Regression - Module - (Housing Prices).csv") #importing second dataset
```

```python
zip_code_condition_sale = pd.pivot_table(data2, index = ["Zipcode"], columns = ['Condition of the House'], values= ["Sale Price"], aggfunc = np.mean) # values say on which column operation is to be performed we grouped it using two different columns here which we have specified in the index and columns section.
zip_code_condition_sale
```

```python
def year(value):
  return value.split()[-1]
# split function divides the string from the part where it encounters space.
# we then stored all that values of years into a seperate column called 'year sold'
data2['year sold'] = data2["Date House was Sold"].map(year)
data2['year sold'].head # printing the column
```

```python
#label encoding 'condtion of the house' column using apply function of pandas library
def label_encoding(haha):
  if haha == 'Fair':
    label=0
  elif haha =='Good':
    label=1
  else:
    label=2
  return label


data2["Condition of the House"] = data2["Condition of the House"].apply(lambda x: label_encoding(x))
data2["Condition of the House"].head()
```
